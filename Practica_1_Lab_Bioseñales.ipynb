*   **Ones:** Es un comando que crea y devuelve una nueva matriz donde todos sus elemetos son cero. Se puede especificar e+la forma y el tipo de datos en sus argmentos de la siguiente manera; *np.ones(forma, tipo de dato, orden: 'C' o 'Fortran')*

*   **Round:** Este comando redondea un n√∫mero o un array de n√∫meros al n√∫mero mas cercano con una cantidad espcifica de decimales. *np.round(numero o array, cant. de decimales)*

```python
# Ejemplo de matriz redondeada
matriz = np.array([ 1.683 ,  2.345 ,  3.857 ,  4.819 ])
matriz_redondeo = np.round(matriz, 2)
print(matriz_redondeo)
```

```python
# Ejemplo de matriz redondeada
matriz_ceil = np.ceil(matriz)
print(matriz_ceil)
```

*   **Floor:** Este comando tambien redondea un numero o array, sin embargo, a diferecia de los anteriores esta redondea el numero o el arreglo de numeros al entreo menor mas cercano.

```python
# Ejemplo de matriz redondeada con comando floor
matriz_floor = np.floor(matriz)
print(matriz_floor)
```
